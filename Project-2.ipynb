# Step 12 Outlier Detection

```python
sns.boxplot(x=df['TotalPrice'],data=df)
plt.title("Outlier Detection")
plt.show()
```

# Step 13 IQR Method

IQR=Q3−Q1

Outlier condition:
x<Q1−1.5(IQR)orx>Q3+1.5(IQR)

```python
Q1=df['TotalPrice'].quantile(0.25)

Q3=df['TotalPrice'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.50 * IQR
upper = Q3 + 1.50 * IQR

Outliers = df[
    (df['TotalPrice'] < lower) | (df['TotalPrice'] > upper)
    ]
Outliers
```
